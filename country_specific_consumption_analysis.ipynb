```python
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle("Electricity consumption per capita across four countries",
            fontweight = "bold",
            fontsize = 20);

ax2.set_yticks(ticks = []);
ax1.set_title("1990")

ax2.set_title("2014")

cmap = "coolwarm" 
# "Blues", Blues, coolwarm, bwr 
# Greens, Reds, binary, copper, viridis,  etc. see https://matplotlib.org/3.5.1/tutorials/colors/colormaps.html

our_world3.plot(column = "1990",
           ax = ax1,
           cmap = cmap,
           edgecolor = "black");

our_world3.plot(column = "2014",
           ax = ax2,
           cmap = cmap,
           edgecolor = "black");

cax = fig.add_axes([0.92, 0.15, 0.03, 0.7]) #colorbar on the right
#[left, bottom, width, height]

sm = plt.cm.ScalarMappable(cmap = cmap,
                           norm = plt.Normalize(vmin = min1, vmax = max1 ))
#scalar mappable

lgd=fig.colorbar(sm, cax=cax)
lgd.set_label(label = "electricity consumption per capita (kWh)")

plt.savefig("Electricity consumption per capita (kWh)", dpi = 300)
```
