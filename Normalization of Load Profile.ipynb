```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as md
import matplotlib.ticker as mtick
```

```python
df = pd.read_excel("normalization.xlsx", skiprows=1, usecols= [0,1,2], index_col= "Time",parse_dates= True)
df
```

```python
profile1_max = df["Profile 1 (kW)"].max()
profile2_max = df["Profile2 (kW)"].max()

print ("Daily Peak (kw) for Profile 1: ", profile1_max)
print ("Daily Peak (kw) for Profile 2: ", profile2_max)
```

```python
df["Normalized Profile 1"] = " "
df["Normalized Profile 2"] = " "
df
```

```python
df["Normalized Profile 1"] = df["Profile 1 (kW)"]/profile1_max
df
```

```python
df["Normalized Profile 2"] = df["Profile2 (kW)"]/profile2_max
df
```

```python
plt.figure(figsize = (12,8))

plt.plot(df.index, df["Profile 1 (kW)"], color = "blue", label = "Profile 1")
plt.plot(df.index, df["Profile2 (kW)"], color = "red", label = "Profile 2")
plt.xlim([df.index[0], df.index[-1]])

plt.xticks(rotation=90)

plt.gca().xaxis.set_major_formatter(md.DateFormatter('%H:%M'))
plt.gca().xaxis.set_major_locator(md.MinuteLocator(byminute=[0,30]))

plt.title("Actual load Profiles (kW)")
plt.legend()
plt.show()

```
